```python
from blackhc import mdp

spec = mdp.MDPSpec()
s0 = spec.state('s0')
s1 = spec.state('s1')
s2 = spec.state('s2')

s3 = spec.state('end', terminal_state=True)

action_set = ['R', 'L', 'B', 'A']
num_actions = len(action_set)
num_agents = 2
actions = []
for i in range(num_actions):
    for j in range(num_actions):
        actions.append(spec.action(action_set[j]+action_set[i]))
print(actions)

#RR
spec.transition(s0, actions[0], mdp.NextState(s0))
spec.transition(s1, actions[0], mdp.NextState(s1))
spec.transition(s2, actions[0], mdp.NextState(s2))
#LR
spec.transition(s0, actions[1], mdp.NextState(s2))
spec.transition(s1, actions[1], mdp.NextState(s0))
spec.transition(s2, actions[1], mdp.NextState(s1))
#BR
spec.transition(s0, actions[2], mdp.NextState(s1))
spec.transition(s1, actions[2], mdp.NextState(s2))
spec.transition(s2, actions[2], mdp.NextState(s0))
spec.transition(s0, actions[2], mdp.Reward(-0.5))
spec.transition(s1, actions[2], mdp.Reward(-0.5))
spec.transition(s2, actions[2], mdp.Reward(-.5))
#AR
spec.transition(s0, actions[3], mdp.NextState(s1))
spec.transition(s1, actions[3], mdp.NextState(s3))
spec.transition(s2, actions[3], mdp.NextState(s0))
spec.transition(s0, actions[3], mdp.Reward(-1.))
spec.transition(s1, actions[3], mdp.Reward(-1.))
spec.transition(s2, actions[3], mdp.Reward(1.))
#RL
spec.transition(s0, actions[4], mdp.NextState(s1))
spec.transition(s1, actions[4], mdp.NextState(s2))
spec.transition(s2, actions[4], mdp.NextState(s0))
#LL
spec.transition(s0, actions[5], mdp.NextState(s0))
spec.transition(s1, actions[5], mdp.NextState(s1))
spec.transition(s2, actions[5], mdp.NextState(s2))
#BL
spec.transition(s0, actions[6], mdp.NextState(s2))
spec.transition(s1, actions[6], mdp.NextState(s0))
spec.transition(s2, actions[6], mdp.NextState(s1))
spec.transition(s0, actions[6], mdp.Reward(-0.5))
spec.transition(s1, actions[6], mdp.Reward(-0.5))
spec.transition(s2, actions[6], mdp.Reward(-0.5))
#AL
spec.transition(s0, actions[7], mdp.NextState(s3))
spec.transition(s1, actions[7], mdp.NextState(s0))
spec.transition(s2, actions[7], mdp.NextState(s1))
spec.transition(s0, actions[7], mdp.Reward(-1.))
spec.transition(s1, actions[7], mdp.Reward(1.))
spec.transition(s2, actions[7], mdp.Reward(-1.))
#RB
spec.transition(s0, actions[8], mdp.NextState(s2))
spec.transition(s1, actions[8], mdp.NextState(s0))
spec.transition(s2, actions[8], mdp.NextState(s1))
spec.transition(s0, actions[8], mdp.Reward(0.5))
spec.transition(s1, actions[8], mdp.Reward(0.5))
spec.transition(s2, actions[8], mdp.Reward(0.5))
#LB
spec.transition(s0, actions[9], mdp.NextState(s1))
spec.transition(s1, actions[9], mdp.NextState(s2))
spec.transition(s2, actions[9], mdp.NextState(s0))
spec.transition(s0, actions[9], mdp.Reward(0.5))
spec.transition(s1, actions[9], mdp.Reward(0.5))
spec.transition(s2, actions[9], mdp.Reward(0.5))
#BB
spec.transition(s0, actions[10], mdp.NextState(s0))
spec.transition(s1, actions[10], mdp.NextState(s1))
spec.transition(s2, actions[10], mdp.NextState(s2))
#AB
spec.transition(s0, actions[11], mdp.NextState(s0))
spec.transition(s1, actions[11], mdp.NextState(s1))
spec.transition(s2, actions[11], mdp.NextState(s2))
spec.transition(s0, actions[11], mdp.Reward(0.5))
#RA
spec.transition(s0, actions[12], mdp.NextState(s3))
spec.transition(s1, actions[12], mdp.NextState(s0))
spec.transition(s2, actions[12], mdp.NextState(s1))
spec.transition(s0, actions[12], mdp.Reward(1.))
spec.transition(s1, actions[12], mdp.Reward(-1.))
spec.transition(s2, actions[12], mdp.Reward(1.))
#LA
spec.transition(s0, actions[13], mdp.NextState(s1))
spec.transition(s1, actions[13], mdp.NextState(s3))
spec.transition(s2, actions[13], mdp.NextState(s1))
spec.transition(s0, actions[13], mdp.Reward(1.))
spec.transition(s1, actions[13], mdp.Reward(1.))
spec.transition(s2, actions[13], mdp.Reward(-1.))
#BA
spec.transition(s0, actions[14], mdp.NextState(s0))
spec.transition(s1, actions[14], mdp.NextState(s1))
spec.transition(s2, actions[14], mdp.NextState(s2))
spec.transition(s0, actions[14], mdp.Reward(-0.5))
#AA
spec.transition(s0, actions[15], mdp.NextState(s3))
spec.transition(s1, actions[15], mdp.NextState(s3))
spec.transition(s2, actions[15], mdp.NextState(s3))
```
